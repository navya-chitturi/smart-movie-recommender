```python
import pandas as pd

# Load ratings
ratings = pd.read_csv('u.data', sep='\t', names=['user_id', 'movie_id', 'rating', 'timestamp'])

# Load movie metadata
movies = pd.read_csv('u.item', sep='|', encoding='latin-1', names=[
    'movie_id', 'title', 'release_date', 'video_release_date', 'IMDb_URL',
    'unknown', 'Action', 'Adventure', 'Animation', 'Children', 'Comedy', 'Crime',
    'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
    'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'
], usecols=range(24))

ratings.head()
```

```python
# Merge ratings and movie metadata on 'movie_id'
df = pd.merge(ratings, movies, on='movie_id')

# Show first 5 rows
df.head()
```

```python
# Drop unneeded columns
df_cleaned = df.drop(columns=['timestamp', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown'])

# Save the cleaned data to a new CSV file
df_cleaned.to_csv('cleaned_movies.csv', index=False)

print("✅ Cleaned dataset saved as cleaned_movies.csv")
```

```python
# Create a user-movie ratings matrix
user_movie_ratings = df_cleaned.pivot_table(index='user_id', columns='title', values='rating')

# Show part of the matrix
user_movie_ratings.head()
```

```python
# Compute correlation between movies
movie_similarity = user_movie_ratings.corr(method='pearson', min_periods=50)

# See similarity with a sample movie
movie_similarity['Titanic (1997)'].dropna().sort_values(ascending=False).head()
```

```python
def get_similar_movies(movie_name, rating_given):
    similar_scores = movie_similarity[movie_name] * (rating_given - 2.5)
    similar_scores = similar_scores.sort_values(ascending=False)
    return similar_scores
```
